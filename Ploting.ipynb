```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("saemark.csv")
df
```

```python
df.corr()
```

```python
plt.figure(figsize=(4,3))
sns.heatmap( df.corr())
```

```python
plt.figure(figsize=(4,3))
sns.heatmap( df.corr(),annot=True)
```

```python
plt.figure(figsize=(4,3))
sns.heatmap( df.corr(),annot=True,cmap='cool')
```

```python
plt.figure(figsize=(4,3))
sns.heatmap( df.corr(),annot=True,cmap='cool',linewidth=10)
```

```python
plt.figure(figsize=(4,3))
sns.heatmap( df.corr(),annot=True,cmap='cool',linewidth='10',linecolor='k')
```
